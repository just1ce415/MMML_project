```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
def inv_t_pdf(pdf):
    abs_val = np.sqrt((np.pi*pdf)**(-1) - 1)
    return abs_val
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
df = 1

norm_rv = stats.norm(mu, sigma)
t_rv = stats.t(df)
x = np.linspace(-20, 20, 1000)
p_norm = 3.2
p_norm_pdf = norm_rv.pdf(p_norm)
p_t = inv_t_pdf(p_norm_pdf)

fig, ax = plt.subplots(1, 1, figsize=(6, 3))

ax.plot(x, norm_rv.pdf(x), color='blue', label='Normal')
ax.plot(x, t_rv.pdf(x), color='orange', label='Student')
ax.axvline(x=p_norm, color='blue')
ax.axvline(x=p_t, color='orange')
ax.axhline(y=p_norm_pdf, color='purple')
ax.scatter([p_norm, p_t], [p_norm_pdf, p_norm_pdf], c='red', marker='o')

ax.set(ylim=(-0.01, 0.04))

plt.legend()
plt.show()
```
